```python
# import libraries 

from bs4 import BeautifulSoup
import requests
import time
import datetime

import smtplib
```

```python
names=[]
authors=[]
ratings=[]
reviews=[]
types=[]
current_price=[]
first_prices=[]
book_urls=[]
asins=[]

keyword='SQL+book'
pages=[str(i)for i in range(0,8)]
for page in pages:
    URL = 'https://www.amazon.com/s?k='+keyword+'&page='+page+'&crid=GNW0L9FR9XFO&qid=1663307970&sprefix=sql+b%2Caps%2C498&ref=sr_pg_'+page

    headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.108 Safari/537.36", "Accept-Encoding":"gzip, deflate", "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8", "DNT":"1","Connection":"close", "Upgrade-Insecure-Requests":"1"}

    time.sleep(10)
    
    response = requests.get(URL, headers=headers)
    if response.status_code != 200:
                print(response.text)
                break

    soup1 = BeautifulSoup(response.content, "html.parser")

    soup2 = BeautifulSoup(soup1.prettify(), "html.parser")
    


    book_containers=soup2.find_all('div',{'data-component-type':'s-search-result'})


   
    for container in book_containers:

        title=container.h2.text.strip()
        names.append(title)

        author=container.find('div',class_='a-row a-size-base a-color-secondary').a
        if author:
            author=container.find('div',class_='a-row a-size-base a-color-secondary').a.text.strip()
            authors.append(author)
        else:
            author=container.find('div',class_='a-row a-size-base a-color-secondary').find_all('span')[1].text.strip()
                
            authors.append(author)
        
            
        
            

        rating_num=rating_num=container.find('div', class_='a-row a-size-small')
        if rating_num:

            rating_num=container.find('div', class_='a-row a-size-small').span.text.strip()[:4]
            ratings.append(float(rating_num))
        else:
            ratings.append(None)


        review=container.find('div',class_='a-section a-spacing-none a-spacing-top-micro').find_all('span')[-1]
        if review:

            review=container.find('div',class_='a-section a-spacing-none a-spacing-top-micro').find_all('span')[-1].text.strip()
            reviews.append(review)
        else:
            reviews.append(None)

        format_book=container.find_all('div',class_='a-row a-size-base a-color-base')[0]
        if format_book:

            format_book=container.find_all('div',class_='a-row a-size-base a-color-base')[0].text.strip()
            types.append(format_book)
        else:
            types.append(None)

        price=container.find('span', class_='a-offscreen')
        if price:
            price=container.find_all('span', class_='a-offscreen')[0].text.strip()
            current_price.append(price)
        else:
            current_price.append(None)

        first_price=container.find_all('span',class_='a-offscreen')
        if len(first_price)>1:
            first_price=container.find_all('span', class_='a-offscreen')[1].text.strip()
            first_prices.append(first_price)
        else:
            first_prices.append(price)


        #find the book's URL
        link=container.find('h2').a
        url=link.get('href')
        book_url = 'https://www.amazon.com/'+url
        book_urls.append(book_url)

        #find the asin
        asin=container.attrs['data-asin']
        asins.append(asin)
```

```python
import pandas as pd
df=pd.DataFrame({'name':names,
'author':authors,
'rating':ratings,
'total_review':reviews,
'type':types,
'current_price':current_price,
'first_price':first_prices,
'book_url':book_urls,
'asin':asins
                })
df.head(10)
```
